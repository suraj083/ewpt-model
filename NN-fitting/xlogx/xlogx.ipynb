```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 

import numpy as np
import tensorflow as tf
import elvet
import matplotlib.pyplot as plt
```

```python
model = elvet.nn(1, 100, 100, 100, 100, 1)
```

```python
x1 = elvet.box((-1, 1, 41)) 

x2 = elvet.box((-5, 5, 201))

x3 = elvet.box((-10, 10, 401))

x4 = elvet.box((-20, 20, 801))

x5 = elvet.box((-50, 50, 2001))

x6 = elvet.box((-100, 100, 4001))
```

```python
def xlogx(x):
    xc = tf.cast(x, tf.complex64)
    return tf.math.real(tf.math.xlogy(xc, xc))
```

```python
def xlogx_deriv(x):
    xb = tf.cast(x, tf.float32)

    with tf.GradientTape() as tape:
        tape.watch(xb) 
        yb = xlogx(xb)
    return tape.gradient(yb, xb)
```

### First go

```python
fitter_xlogx = elvet.fitter(x1, xlogx(x1), model=model, epochs=50000, lr=0.0007)
```

```python
plt.subplot(2,3,1)
plt.plot(x1, fitter_xlogx.model(x1))
plt.plot(x1, xlogx(x1), linestyle="dotted")

plt.subplot(2,3,2)
plt.plot(x2, fitter_xlogx.model(x2))
plt.plot(x2, xlogx(x2), linestyle="dotted")

plt.subplot(2,3,3)
plt.plot(x3, fitter_xlogx.model(x3))
plt.plot(x3, xlogx(x3), linestyle="dotted")

plt.subplot(2,3,4)
plt.plot(x4, fitter_xlogx.model(x4))
plt.plot(x4, xlogx(x4), linestyle="dotted")

plt.subplot(2,3,5)
plt.plot(x5, fitter_xlogx.model(x5))
plt.plot(x5, xlogx(x5), linestyle="dotted")

plt.subplot(2,3,6)
plt.plot(x6, fitter_xlogx.model(x6))
plt.plot(x6, xlogx(x6), linestyle="dotted")
```

```python
def fitter_xlogx_deriv(x):
    xb = tf.cast(x, tf.float32)

    with tf.GradientTape() as tape:
        tape.watch(xb) 
        yb = fitter_xlogx.model(xb)
    return tape.gradient(yb, xb)
```

```python
plt.subplot(2,3,1)
plt.plot(x1, fitter_xlogx_deriv(x1))
plt.plot(x1, xlogx_deriv(x1), linestyle="dotted")

plt.subplot(2,3,2)
plt.plot(x2, fitter_xlogx_deriv(x2))
plt.plot(x2, xlogx_deriv(x2), linestyle="dotted")

plt.subplot(2,3,3)
plt.plot(x3, fitter_xlogx_deriv(x3))
plt.plot(x3, xlogx_deriv(x3), linestyle="dotted")

plt.subplot(2,3,4)
plt.plot(x4, fitter_xlogx_deriv(x4))
plt.plot(x4, xlogx_deriv(x4), linestyle="dotted")

plt.subplot(2,3,5)
plt.plot(x5, fitter_xlogx_deriv(x5))
plt.plot(x5, xlogx_deriv(x5), linestyle="dotted")

plt.subplot(2,3,6)
plt.plot(x6, fitter_xlogx_deriv(x6))
plt.plot(x6, xlogx_deriv(x6), linestyle="dotted")
```
